### Looping Through Entire List

Often we want to run through all entries in a list to person the same task with each item. For looping we can use a for loop. Heres an example of how you could use for loop on a list:

```python
names = ['adil', 'ali', 'abdullah']
for name in names:
    print(name)
```



Unlike other languages like C/C++ Python does not have curly brackets to define the body of loops. Instead Python uses indentation to know whether the lines following a loop is within the loops body or outside it. So its important to keep in mind the indentations of your code while writing your loops. Another thing to keep in mind is the colon : that comes after the loop statement. Without that the you'll get a syntax error.

### The Range Function

Pythons range() function makes it easy to generate a series of numbers.

```python
for value in range(1, 5):
    print(value)
```

the range function acccepts 2 arguments, the starting value and the ending value. As you can notice the starting value is inclusive and the ending value is exclusive in the range function.

We can use this range function to create a numeric list as well

```python
numbers = list(range(1,5))
print(numbers)
```




### List Comprehensions

A list comprehension allows you to combine a for loop and the creation of new elements into one line and automatically appends the new element to the end of the list.

```python
squares = [value**2 for value in range(1,11)]
print(squares)
```

## Working with Part of a List

You sometimes may need to divide the data you have in a list into different groups. For this you'll need to slice the data to create different groups

### Slicing a List

Slicing lists require you to define the starting index and ending index. Similar to the range function the starting value is inclusive to the slice while the ending value is exclusive to the slice

```python
players = ['charles', 'martina', 'micheal', 'florence', 'eli']
print(players[0:3])
```

## Tuples

A tuble looks just like a list except you use parentheses instead of square brackets. Tuples are just an immutable list, meaning you cannot access its index and reassign it with some new data.
