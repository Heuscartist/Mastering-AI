## Chapter#6 : Dictionaries

In this chapter we will learn how to use dictionaries which allow us to store information that is related.

We'll start of by creating a simple dictionary about ourselves. This dictionary will contain our basic information and will give us an idea of the basic syntax of how dictionaries are created

```python
person = {'name':'adil', 'age':23, 'hobby':'gaming'}
print(person)
```

A dictionary is a collection of key-value pairs. Each key is connected to a value and we can access the value using its associated key.

### Accessing Values in  Dictionary

To access the values within a dictionary we need to pass in the key associated with the value we want to retrieve.

```python
print(person['name'])
```

sometimes we need to access values from dictionary but we are unsure of the keys that are available in the dict. We can use the keys() method on the dictionary to print the available keys in a dictionary

```python
print(person.keys())
```

### Adding Values in Dictionaries

Sometimes we'll need to add more information in the dictionary. For this we'll need to create new keys and assign new values to those

```python
person['favourite food'] = 'pizza'
print(person)
```

### Modifying Values associated with a key

similary we can also change the values associated with a key in a similar syntax

```python
person['favourite food'] = 'karahi'
print(person)
```

### Looping Through All Key-Value Pairs

Loops can be used to access all the the items in a dictionary

```python
for key, value in person.items():
    print("\nKey: " , key)
    print("Value:" , value)
```
