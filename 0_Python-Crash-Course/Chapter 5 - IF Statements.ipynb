## Chapter#5 : IF Statements

In this chapter we will learn how to write condition tests and check our condition of interest. Then we will apply these concepts to lists

We will use a simple example to understand conditioning in Python.

```python
cars = ['bmw', 'audi', 'porche']
for car in cars:
    if car == 'bmw':
        print(car.upper())
    else:
        print(car.title())
```

the above example shows the syntax of the if-else statement. Similar to for loops we have a colon : after the conditional statement and the following code that is indented is considered as part of the statements body. Multiple conditions can be checked using the and / or keywords. For Example:

### The if-elif-else Chain

Often there will be situations where there are multiple conditions to check and you will need to execute code depending on these conditions. For this the if-elif-else chain is used. The elif statement takes a condition in its statement similar to the if statement.

```python
age = 12
if age < 4:
    print("Your admission costs $0")
elif age < 18:
    print("Your admission costs $5")
else:
    print("Your admission costs $10")
     
```
